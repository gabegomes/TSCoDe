```python
from cclib.io import ccread
from ase import Atoms
from ase.visualize import view
import os
import numpy as np
os.chdir('Resources/SN2')
mol = ccread('TS_out_test.xyz')

coords = mol.atomcoords[1]
atoms = Atoms(mol.atomnos, positions=coords)
# view(atoms)
constraints = np.array([[2,13], [7,15]])
```

```python
from simtk.openmm import *
from simtk.openmm.app import *
from simtk.unit import *

os.system('obabel TS_out_test.xyz -o pdb -O temp.pdb')
pdb = PDBFile('temp.pdb')
forcefield = ForceField('amber99sb.xml')

system = forcefield.createSystem(pdb.topology, nonbondedMethod=PME, nonbondedCutoff=1.0*nanometer, constraints=HBonds)
integrator = LangevinIntegrator(300*kelvin, 1/picosecond, 0.002*picoseconds)

simulation = Simulation(pdb.topology, system, integrator)
simulation.context.setPositions(pdb.positions)

simulation.minimizeEnergy(maxIterations=100)

pdb.positions
```
